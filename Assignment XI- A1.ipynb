###**Method 1: Naive Method**

*   In this method, one just runs a loop and increases the counter till the last element of the list to know its count. This is the most basic strategy that can be possibly employed in the absence of other present techniques.




```python
from collections import Counter
list=[1,4,5,6,7,8]
print("The list is :" + str(list))

# Finding length of list
# using loop
# Initializing counter
counter = 0
for i in list:
  counter = counter + 1

print("Length of list using navie mathod is:" + str(counter))
```

#**Q.15 You are given a list of integer elements. Make a new list that will store a square of elements of the previous list. (With and without list comprehension)**

i. Input_list = [2,5,6,12]

ii. Output_list = [4,25,36,144]

```python
input_list = [2,5,6,12]

using_list_comphression=[i**2 for i in input_list]

print('using_list_comphression {}'.format(using_list_comphression))
```

```python
def square(input_list):
    return [i ** 2 for i in input_list]

print(f"without list comprehension {square(input_list)}")
```
